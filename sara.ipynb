```python
import gradio as gr
import numpy as np
from tensorflow import keras

model = keras.models.load_model('/content/drive/MyDrive/sara_S8/Projet_IA/Dossier sans titre/traffic_sign_model.h5', compile=False)
class_names = classNames


def predict_image(img):
    # Preprocess the input image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img, (32, 32))
    img = np.expand_dims(img, axis=-1)
    img = np.expand_dims(img, axis=0)
    img = img.astype('float32') / 255.

    # Make predictions using the model
    predictions = model.predict(img)
    class_index = np.argmax(predictions[1])
    class_label = class_names[class_index]

    # Return the predicted class label
    return class_label


inputs = gr.inputs.Image(shape=(None, None))
outputs = gr.outputs.Label(num_top_classes=1)

title = "Traffic Sign Classifier"
description = "A simple web app to classify traffic signs."
gr.Interface(fn=predict_image, inputs=inputs, outputs=outputs, title=title, description=description).launch(share=True)
```
